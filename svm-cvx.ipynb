```python
import cvxpy as cp
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
X = pd.read_csv('X.csv').drop('Unnamed: 0', axis=1)
y = pd.read_csv('y.csv').drop('Unnamed: 0', axis=1)
scaler = StandardScaler(copy=False)
scaler.fit_transform(X);  # first standardize X
X = X.values
y = y.values.reshape(X.shape[0])
```

```python
y = np.sign(y)
```

```python
n = X.shape[0]
p = X.shape[1]

w = cp.Variable(p)
w0 = cp.Variable(1)
ksai = cp.Variable(n)

objective = cp.Minimize(1/2 * cp.sum_squares(w) + 2 * cp.sum(ksai))
constraints = [ksai >= 0, y * (X * w + w0) >= 1 - ksai]
prob = cp.Problem(objective, constraints)

result = prob.solve()

print(w.value)
```
