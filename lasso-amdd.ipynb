```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
X = pd.read_csv('X.csv').drop('Unnamed: 0', axis=1)
y = pd.read_csv('y.csv').drop('Unnamed: 0', axis=1)
scaler = StandardScaler(copy=False)
scaler.fit_transform(X);  # first standardize X
```

```python
X = X.values
y = y.values.reshape(X.shape[0])
```

```python
y = y - y.mean()
```

```python
lam = 20
```

```python
def S_kappa(a, kappa):
    if a > kappa:
        return a - kappa
    elif a < -kappa:
        return a + kappa
    else:
        return 0
vec_S_kappa = np.vectorize(S_kappa, excluded='kappa', otypes=[np.float128])
```
